```python
def Goal_diff_count(x,y):
    laliga['GoalDiff']=x-y
    laliga['GoalDiff']=abs(laliga['GoalDiff'])
    return laliga [laliga['GoalDiff']==laliga['GoalDiff'].max()],laliga [laliga['GoalDiff']==laliga['GoalDiff'].min()]

Goal_diff_count(laliga['GoalsFor'],laliga['GoalsAgainst'])
```

```python
laliga['GamesWon']=pd.to_numeric(laliga['GamesWon'])
laliga['GamesPlayed']=pd.to_numeric(laliga['GamesPlayed'])
laliga['WinningPercent']=laliga['GamesWon']/laliga['GamesPlayed']*100
laliga.fillna(0,inplace= True)
laliga.head()
```
